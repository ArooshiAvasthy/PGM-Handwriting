```python
#Creating a bayesian model by adding nodes x1-x6, adding relations between nodes based intuition and taking in account a
# threshold for the correlations calculated
model1 = BayesianModel()
model1.add_nodes_from(['x1', 'x2', 'x3','x4','x5','x6'])
model1.add_edges_from([('x1', 'x4'), ('x4', 'x2'),('x2','x5'),('x5','x3'),('x3','x6')])
x1_cpd=TabularCPD('x1',4,[[0.780],[0.015],[0.055],[0.15]])
model1.add_cpds(x4x1_cpd,x2x4_cpd,x5x2_cpd,x3x5_cpd,x6x3_cpd,x1_cpd)
#forward sampling is used for generating data for the given model
from pgmpy.sampling import BayesianModelSampling
inference1 = BayesianModelSampling(model1)
dataModel1 = np.array(inference1.forward_sample(size=1000, return_type='recarray'))

```

```python
#Creating a bayesian model by adding nodes x1-x6, adding relations between nodes based intuition and taking in account a
# threshold for the correlations calculated
model2 = BayesianModel()
model2.add_nodes_from(['x1', 'x2', 'x3','x4','x5','x6'])
model2.add_edges_from([('x1', 'x4'), ('x4', 'x6'),('x6','x3'),('x3','x5'),('x5','x2')])
x1_cpd=TabularCPD('x1',4,[[0.780],[0.015],[0.055],[0.15]])
x6_cpd=TabularCPD('x6',5,[[0.15],[0.32],[0.14],[0.18],[0.21]])
model2.add_cpds(x4x1_cpd,x6x4_cpd,x3x6_cpd,x5x3_cpd,x2x5_cpd,x1_cpd)
#forward sampling is used for generating data for the given model
from pgmpy.sampling import BayesianModelSampling
inference2 = BayesianModelSampling(model2)
dataModel2 =np.array(inference2.forward_sample(size=1000, return_type='recarray'))

```

```python
#Creating a bayesian model by adding nodes x1-x6, adding relations between nodes based intuition and taking in account a
# threshold for the correlations calculated
model3 = BayesianModel()
model3.add_nodes_from(['x1', 'x2', 'x3','x4','x5','x6'])
model3.add_edges_from([('x1', 'x2'), ('x2', 'x5'),('x5','x3'),('x3','x6'),('x6','x4')])
x1_cpd=TabularCPD('x1',4,[[0.780],[0.015],[0.055],[0.15]])
#x6_cpd=TabularCPD('x6',5,[[0.15],[0.32],[0.14],[0.18],[0.21]])
model3.add_cpds(x2x1_cpd,x5x2_cpd,x3x5_cpd,x6x3_cpd,x4x6_cpd,x1_cpd)
#forward sampling is used for generating data for the given model
from pgmpy.sampling import BayesianModelSampling
inference3 = BayesianModelSampling(model3)
dataModel3 = np.array(inference3.forward_sample(size=1000, return_type='recarray'))
```

```python
#Concatenating the data generated through model1, model2, model3
data= np.concatenate((dataModel1,dataModel2,dataModel3))
data=pd.DataFrame(data)
print(data)
```

```python
#Implementing Hill Climb Search  and using K2 score over the dataset to find the best model
from pgmpy.estimators import HillClimbSearch

hc = HillClimbSearch(data, scoring_method=K2Score(data))
best_model = hc.estimate()
print(best_model.edges())
```

```python
#Implementing Bayesian model basel on the best model relation calculated using Hill Climb Search
from pgmpy.inference import VariableElimination
bmodel = BayesianModel([('x2', 'x1'), ('x3', 'x4'), ('x3', 'x5'), ('x3', 'x6'), ('x6', 'x5'), ('x6', 'x4'), ('x4', 'x2')])
bmodel.fit(data)
inference = VariableElimination(bmodel)
comp = inference.query(['x2'])['x2']
print(comp)
```

```python
#Bayesian to Markov
mm = bmodel.to_markov_model()
inference_mm = VariableElimination(mm)
comp1 = inference.query(['x2'])['x2']
print(comp1)
```
